#### 基于倾向分的处理方法

```python
model_t.predict_proba(X)
```

```python
testing_treat = model_t.predict_proba(X)
testing_treat = pd.DataFrame(testing_treat)
testing_treat.columns = ['propensity0','propensity1','propensity2','propensity3']

limit=0.05
propensities_clipped = np.clip(testing_treat, a_min=limit, a_max=1 - limit)
weights = 1 / propensities_clipped
# keep sum of weights equal to nr of samples:
weights = weights * len(propensities_clipped) / np.sum(weights)
```

```python
pd_outcome.head(2)
```

```python
pd_outcome["uplift1_ipw"]=pd_outcome["all_treatment=1"] / weights["propensity1"] - pd_outcome["all_treatment=0"] / weights["propensity0"]
pd_outcome["uplift2_ipw"]=pd_outcome["all_treatment=2"] / weights["propensity2"] - pd_outcome["all_treatment=0"] / weights["propensity0"]
pd_outcome["uplift3_ipw"]=pd_outcome["all_treatment=3"] / weights["propensity3"] - pd_outcome["all_treatment=0"] / weights["propensity0"]
```
